# **Seaborn**

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

````python
# Load a sample dataset from seaborn
data = sns.load_dataset("tips")
````

## **1. Basic Plotting**

### **1.1 Relational Plots**

#### **a. Relplot**


````python
sns.relplot(data=data, x="total_bill", y="tip", hue="time")
plt.show()
````

#### **b. Scatterplot**

````python
sns.scatterplot(data=data, x="total_bill", y="tip", hue="time")
plt.show()
````

#### **c. Lineplot**

````python
sns.lineplot(data=data, x="total_bill", y="tip", hue="time")
plt.show()
````

### **1.2 Categorical Plots**

````python
# Catplot
sns.catplot(data=data, x="day", y="total_bill", hue="smoker", kind="swarm")
plt.show()
````

#### **a. Stripplot**

````python
sns.stripplot(data=data, x="day", y="total_bill", hue="smoker", dodge=True)
plt.show()
````

#### **b. Swarmplot**

````python
sns.swarmplot(data=data, x="day", y="total_bill", hue="smoker", dodge=True)
plt.show()
````

####  **c. Boxplot**

````python
sns.boxplot(data=data, x="day", y="total_bill", hue="smoker")
plt.show()
````

#### **d. Violinplot**

````python
sns.violinplot(data=data, x="day", y="total_bill", hue="smoker", split=True)
plt.show()
````

#### **e. Barplot**

````python
sns.barplot(data=data, x="day", y="total_bill", hue="smoker")
plt.show()
````

#### **f. Pointplot**

````python
sns.pointplot(data=data, x="day", y="total_bill", hue="smoker")
plt.show()
````

### **1.3 Distribution Plots**

#### **a. Displot**

````python

sns.displot(data=data, x="total_bill", hue="smoker", kind="kde")
plt.show()
````

#### **b. Histplot**

````python
sns.histplot(data=data, x="total_bill", hue="smoker", multiple="stack")
plt.show()
````

#### **c. KDEplot**

````python
sns.kdeplot(data=data, x="total_bill", hue="smoker", fill=True)
plt.show()
````

#### **d. ECDFplot**

````python
sns.ecdfplot(data=data, x="total_bill", hue="smoker")
plt.show()
````

#### **e. Rugplot**

````python
sns.rugplot(data=data, x="total_bill", hue="smoker")
plt.show()
````

### **1.4 Matrix Plots**

#### **a. Heatmap**

````python
# Create a synthetic DataFrame with random numeric values
n_rows = 100
n_cols = 5
df = pd.DataFrame(np.random.rand(n_rows, n_cols), columns=[f"col_{i}" for i in range(n_cols)])

# Calculate correlation
corr = df.corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
````

#### **b. Clustermap**

````python
# Clustermap
plt.figure(figsize=(8, 6))
sns.clustermap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Clustermap")
plt.show()
````

### **1.5 Regression Plots**

#### **a. Regplot**

````python
sns.regplot(data=data, x="total_bill", y="tip")
plt.show()
````

#### **b. Lmplot**

````python
sns.lmplot(data=data, x="total_bill", y="tip", hue="smoker")
plt.show()
````

#### **c. Residplot**

````python
sns.residplot(data=data, x="total_bill", y="tip")
plt.show()
````

## **2. Plot Customization**

### **2.1 Titles and Labels**

````python
ax = sns.scatterplot(data=data, x="total_bill", y="tip", hue="time")
ax.set_title("Total Bill vs Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
plt.show()
````

### **2.2 Legends**

````python
ax = sns.scatterplot(data=data, x="total_bill", y="tip", hue="time")
ax.legend(title="Time of Day")
plt.show()
````

### **2.3 Axes**

````python
ax = sns.scatterplot(data=data, x="total_bill", y="tip", hue="time")
ax.set_xlim(0, 60)
ax.set_ylim(0, 12)
plt.show()
````

### **2.4 Grids**

````python
sns.set(style="whitegrid")
ax = sns.scatterplot(data=data, x="total_bill", y="tip", hue="time")
plt.show()

sns.despine()
plt.show()
````

### **2.5 Faceting**

#### **a. FacetGrid**

````python
g = sns.FacetGrid(data, col="time")
g.map(sns.histplot, "total_bill")
plt.show()
````

#### **b. PairGrid**

````python
g = sns.PairGrid(data)
g.map(sns.scatterplot)
plt.show()
````

## **3. Styling**

#### **3.1 Themes**

````python
sns.set_theme(style="darkgrid")
sns.scatterplot(data=data, x="total_bill", y="tip", hue="time")
plt.show()

sns.set_context("notebook")
sns.scatterplot(data=data, x="total_bill", y="tip", hue="time")
plt.show()
````

#### **3.2 Color palettes**

````python
sns.set_palette("pastel")
sns.scatterplot(data=data, x="total_bill", y="tip", hue="time")
plt.show()

sns.palplot(sns.color_palette("viridis"))
plt.show()
````

#### **3.3 Aesthetics**

````python
sns.set_style("white")
sns.scatterplot(data=data, x="total_bill", y="tip", hue="time")
plt.show()

sns.axes_style()
plt.show()

sns.plotting_context()
plt.show()
````

#### **4.3 Pair Grids**

````python
g = sns.PairGrid(data, hue="smoker")
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
plt.show()
````

````python
g = sns.FacetGrid(data, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
````
